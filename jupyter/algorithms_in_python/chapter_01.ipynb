## R-1.3

```python
def minmax(data):
    if not len(data):
        return (None, None)
    
    min = data[0]
    max = data[0]
    
    for n in data:
        if min > n:
            min = n
        if max < n:
            max = n
    
    return (min, max)

for data in ((2, 3, 5, 1, 7, 2, 3), (), (2, 2, 2, 2)):
    print(data)
    print(minmax(data))
    print()
```

## R-1.4

```python
def sumsq_smaller(n):
    sum = 0
    for i in range(1, n):
        sum += i * i
    
    return sum

for n in range(11):
    print('%d: %d' % (n, sumsq_smaller(n)))
```

## R-1.6

```python
def sumsq_even_smaller(n):
    sum = 0
    for i in range(1, n):
        if i % 2:
            sum += i * i
    return sum

for n in range(11):
    print('%d: %d' % (n, sumsq_even_smaller(n)))
```

## R-1.12

```python
import random

def choice(li):
    if not li:
        return None
    return li[random.randrange(len(li))]

for li in [[1, 2, 3, 4], [0], []]:
    print(li)
    print(choice(li))
```

## P-1.30

```python
def count_div_2(n):
    count = 0
    
    while n >= 2:
        n /= 2
        count += 1
    
    return count
    
for n in (3, 10, 100):
    print('%d: %d' % (n, count_div_2(n)))
```

## P-1.31

```python
mon_values = [100.0, 20.0, 10.0, 5.0, 1.0, 0.25, 0.1]

def make_change(value):
    mon_counts = [0] * len(mon_values)
    
    for n in range(len(mon_values)):
        mon_value = mon_values[n]
        count = value / mon_value
        count = int(count)
        
        value -= mon_value * count
        
        mon_counts[n] = count
    return mon_counts

for value in (125.2, 10.2, 3.7, 1024.7):
    print('%f: ' % value)
    print(make_change(value))
    print()
```

## P-1.34

```python
import random

sentence = 'I will never spam my friends again.'

def typo_pos(length):
    pos_list = []
    
    while len(pos_list) < 8:
        rand_pos = random.randrange(length)
        if rand_pos in pos_list:
            continue
        pos_list.append(rand_pos)
    
    return pos_list
        
        
def print_punish():
    pos_list = typo_pos(len(sentence))
    
    new_sentence = sentence
    for pos in pos_list:
        rand_char = chr(random.randrange(ord('a'), ord('z') + 1))
        new_sentence = new_sentence[:pos] + rand_char + new_sentence[pos+1:]
    
    print(new_sentence)
    
for n in range(15):
    print_punish()
    
```

## P-1.35

```python
import random

def check_same_birth(n):
    birthday_list = []
    for n in range(n):
        birthday_list.append(random.randrange(365))
                             
    if len(birthday_list) != len(set(birthday_list)):
        return True
    
    return False

def calc_chance(n, times):
    same_count = 0
    
    for nn in range(times):
        if check_same_birth(n):
            same_count += 1
    
    return same_count

for n in range(5, 101, 5):
    print('%d: %d%%' % (n, calc_chance(n, 100)))
                             
```

## P-1.36

```python
def count_overlap(sentence):
    d = {}
    
    words = sentence.split(' ')
    
    for word in words:
        if word in d.keys():
            d[word] += 1
        else:
            d[word] = 1
    
    return d

d = count_overlap('hello rabbit howdi rabbit howdi there hello there')
d
```
